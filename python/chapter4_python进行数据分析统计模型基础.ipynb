# 统计模型
由于一般的概率论难以描述现实中的现象。比如：食堂的菜好吃吗？那么明天该进什么货，那菜市场的价格变了很多怎么办呢？
那么，使用现实世界对应的模型有助于我们`理解和预测现实事物`。

## 一些概念
数学模型：用数学式表达
概率模型：用概率的语言表达

那么统计模型：基于数据建立的概率模型。使用了概率论，并且经过了修正，那么也就是说更贴近实际。

统计模型让数据分析有了很大的进步，勘称现代数据分析的标准基础。

## 建模方法

### 响应变量和解释变量等术语
响应变量：根据某个因素而变化（相应）的变量。自变量
解释变量：对关注的对象的变化进行解释的变量。因变量

概率模型多记作：相应变量 ~ 解释变量。

参数模型：用尽量少的参数表达的模型
非参数模型：则相反了。

线性模型，顾名思义，即对于一类可以用`矩阵形式`进行表达的模型的统称

### 系数和权重
模型中使用的参数叫做系数
在机器学习中，模型的系数也叫做`权重`

### 建模 = 模型选择 + 参数估计
变量的选择：为模型选择解释变量

#### 那么一般可分为
-   通过假设-检验选择变量
-   通过信息准则(Akaike Information Criterion, AIC)选择变量
    -   AIC越小，模型越合适。

### 模型评估
评估模型的方法有两种：
-   评估预测精度，精度越高，越准
-   检验模型是否满足建模时所假设的条件







## 数据表示和模型名称

### 正态线性模型
假设响应变量服从正态分布的线性模型。

#### 术语
回归分析：解释变量为连续变量的模型。

多元回归分析：含有多个解释变量的

方差分析：解释变量为分类变量的模型。

那么不符合正态分布的线性模型，叫做广义线性模型。
